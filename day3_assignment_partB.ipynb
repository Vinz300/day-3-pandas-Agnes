```python
import yfinance as yf
```

```python
import pandas as pd
```

```python
#create Pandas Series for Gold and Nifty Prices 
gold=yf.Ticker("GLD").history(start='2023-03-13',end='2023-04-13')
gold
```

```python
nifty=yf.Ticker("^NSEI").history(start='2023-03-13',end='2023-04-13')
nifty
```

```python
# Print the first 5 prices for each and their data types
print("first five gold prices:")
print(gold.head(5))
print("datatype:", type(gold))
```

```python
print("first five nifty prices:")
print(nifty.head(5))
print("datatype:", type(nifty))
```

```python
#check lenths and align byt dates
print("lenth before aligning:")
print("gold:", len(gold))
print("nifty:", len(nifty))
dates=gold.index.intersection(nifty.index)
gold_aligned=gold.loc[dates]
nifty_aligned=nifty.loc[dates]
print("length after aligning:")
print("gold", len(gold_aligned))
print("nifty", len(nifty_aligned))
```

```python
#Compute daily percentage returns for Nifty prices
nifty_percentage=nifty_aligned.pct_change()*100
print("NIFTY daily percemtage for first five:")
print(nifty_percentage.dropna().head(5)) 
```

```python
#correlation between gold and NIFTY
combined_df=pd.concat([gold_aligned,nifty_aligned],axis=1)
Correlation=combined_df.corr().iloc[0,1]
print("Correlation between gold and nifty is",Correlation)
```
